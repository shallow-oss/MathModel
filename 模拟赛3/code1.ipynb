```python
import pandas as pd
import numpy as np
```

```python
# 读取Excel文件
data = pd.read_excel('附件2 慢性病及相关因素流调数据.xlsx')
data
```

```python
# 获取完整的列名列表
column_names = list(data.columns)
# 将列名保存到文件
with open('column_names.txt', 'w', encoding='utf-8') as file:
    for name in column_names:
        file.write(name + '\n')
```

```python
data.isnull().sum()
# 获取完整的列名列表
null_column_names = list(data.isnull().sum())
# 将列名保存到文件
with open('null_column_names.txt', 'w', encoding='utf-8') as file:
    for name,none in zip(column_names,null_column_names):
        file.write(name + ' : ' + str(none) + '\n')
```
