```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
x = np.random.rand(N) *6 - 3
x.sort()
```

```python
y = np.sin(x) + np.random.rand(N) * 0.05
```

```python
x = x.reshape(-1,1)
```

```python
dt_reg = DecisionTreeRegressor(criterion='friedman_mse',max_depth=5)
dt_reg.fit(x,y)
```

```python
x_test = np.linspace(-3,3,50).reshape(-1,1)
y_hat = dt_reg.predict(x_test)
```

```python
plt.plot(x,y,'y*',label='actual')
plt.plot(x_test,y_hat,'b-',lw=2,label='pred')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
deepth = [2,4,6,8,10]
color = 'rgbmy'
plt.plot(x,y,'ko',label='actual')
x_test = np.linspace(-3,3,50).reshape(-1,1)
for d,c in zip(deepth,color):
    dt_reg = DecisionTreeRegressor(criterion='friedman_mse',max_depth=d)
    dt_reg.fit(x,y)
    y_hat = dt_reg.predict(x_test)
    plt.plot(x_test,y_hat,'-',lw=2,label='depth %d' % d,color=c )
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
