```python
import pandas as pd
import numpy as np
```

```python
soil_moisture=pd.read_excel('data\附件3、土壤湿度2022—2012年.xls')
soil_moisture=soil_moisture[['月份','年份','10cm湿度(kg/m2)','40cm湿度(kg/m2)',	'100cm湿度(kg/m2)',	'200cm湿度(kg/m2)']]
soil_moisture.head()
```
